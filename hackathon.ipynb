# converting column data into numeric data using LabelEncoder

```python
df.columns
```

```python
x=['age', 'workclass', 'fnlwgt', 'education', 'education-num',
       'marital-status', 'occupation', 'relationship', 'race', 'sex',
       'capital-gain', 'capital-loss', 'hours-per-week', 'country', 'salary']

for i in x:
    df[i]=la.fit_transform(df[i])
```

```python
df
```

```python
from sklearn.linear_model import LogisticRegression
lor=LogisticRegression()
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1:].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=40)
```

```python
lor.fit(x_train,y_train)
```

```python
y_pred = lor.predict(x_test)
pd.DataFrame({'Actual ': list(y_test),'predicted ': list(y_pred)})
```

```python
from sklearn.metrics import accuracy_score,f1_score
print('Actual : ',accuracy_score(y_test,y_pred)*100,'%')
print('F1_score : ',f1_score(y_test,y_pred))
```
